```python
# Paso 1: Filtrar solo medallistas
medallistas = data[data['Medalla'] != 'Sin medalla']

# Paso 2: Agrupar por año y equipo
medallas_por_equipo = medallistas.groupby(['Año', 'Equipo']).size().reset_index(name='Conteo')

# Paso 3: Pivotear datos
pivot = medallas_por_equipo.pivot(index='Año', columns='Equipo', values='Conteo').fillna(0)

# Paso 4: Graficar top 5 equipos con más medallas
top_5_equipos = pivot.sum().sort_values(ascending=False).head(5).index
pivot[top_5_equipos].plot(figsize=(12, 6))

plt.title('Evolución de medallas por equipo')
plt.xlabel('Año')
plt.ylabel('Número de medallas')
plt.legend(title='Equipo')
plt.grid(True)
plt.show()
```

```python


# Paso 1: Filtrar medallistas
medallistas = data[data['Medalla'] != 'Sin medalla']

# Paso 2: Filtrar últimos 30 años
ultimo_anio = medallistas['Año'].max()
anio_limite = ultimo_anio - 30
medallistas_reciente = medallistas[medallistas['Año'] >= anio_limite]

# Paso 3: Agrupar por año y equipo
medallas_por_equipo = medallistas_reciente.groupby(['Año', 'Equipo']).size().reset_index(name='Conteo')

# Paso 4: Pivotear datos
pivot = medallas_por_equipo.pivot(index='Año', columns='Equipo', values='Conteo').fillna(0)

# Paso 5: Graficar top 5 equipos
top_5_equipos = pivot.sum().sort_values(ascending=False).head(5).index
pivot[top_5_equipos].plot(figsize=(12, 6))

plt.title(f'Evolución de medallas por equipo (últimos 30 años)')
plt.xlabel('Año')
plt.ylabel('Número de medallas')
plt.legend(title='Equipo')
plt.grid(True)
plt.show()
```

```python


# Filtrar medallistas y últimos 30 años
medallistas = data[data['Medalla'] != 'Sin medalla']
ultimo_anio = medallistas['Año'].max()
anio_limite = ultimo_anio - 30
medallistas_reciente = medallistas[medallistas['Año'] >= anio_limite]

# Agrupar por año y equipo
medallas_por_equipo = medallistas_reciente.groupby(['Año', 'Equipo']).size().reset_index(name='Conteo')

# Pivotear: filas=años, columnas=equipos, valores=conteo
pivot = medallas_por_equipo.pivot(index='Año', columns='Equipo', values='Conteo').fillna(0)

# Elegimos los 5 equipos más fuertes del periodo
top_equipos = pivot.sum().sort_values(ascending=False).head(5).index
pivot_top = pivot[top_equipos]

# Graficar barras apiladas
pivot_top.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab10')

plt.title('Evolución de medallas por equipo (últimos 30 años)')
plt.xlabel('Año')
plt.ylabel('Número de medallas')
plt.legend(title='Equipo', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python

# 1. Filtrar solo medallistas
medallistas = data[data['Medalla'] != 'Sin medalla']

# 2. Agrupar: país y tipo de medalla
medallas_pais = medallistas.groupby(['Equipo', 'Medalla']).size().unstack(fill_value=0)

# 3. Calcular total medallas por país y ordenar
medallas_pais['Total'] = medallas_pais.sum(axis=1)
top10_paises = medallas_pais.sort_values('Total', ascending=False).head(10)

# 4. Gráfico de barras apiladas verticales
top10_paises[['Gold', 'Silver', 'Bronze']].plot(
    kind='barh', stacked=True, figsize=(10, 6),
    color=['gold', 'silver', '#cd7f32']  # colores oro, plata, bronce
)

# 5. Estética
plt.title('Top 10 países con más medallas')
plt.xlabel('Cantidad de medallas')
plt.ylabel('País')
plt.legend(title='Tipo de Medalla')
plt.gca().invert_yaxis()  # Para que el país con más medallas quede arriba
plt.tight_layout()
plt.show()
```
